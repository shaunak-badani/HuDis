```python
df5_file = 'weights.hdf5'  # Replace with your HDF5 file path
def open_group(f):
    # Print the top-level keys (names of groups or datasets)
    print("Keys in the HDF5 file:", list(f.keys()))

    # Iterate over the top-level groups or datasets
    for key in f.keys():
        print(f"\nKey: {key}")

        # Check if it's a dataset or group
        item = f[key]

        if isinstance(item, h5py.Group):
            print(f"  {key} is a group.")
            # If it's a group, list its contents
            open_group(item)
        elif isinstance(item, h5py.Dataset):
            print(f"  {key} is a dataset.")
            # Print the data stored in the dataset
            print(f"  Data: {item[:]}")  # Access the whole dataset (as numpy array)
            print(f"  Data shape: {item.shape}")  # Print the shape of the dataset
            print(f"  Data dtype: {item.dtype}")  # Print the dtype of the datase
```

```python
with h5py.File(hdf5_file, 'r') as f:
    open_group(f)
```
